# Variable Discretization
---

Some Machine Learning technique require categorical variables, so we have to transform or __discretize__ the data.

There are two types of discretization: __Supervisioned__ and __Unsupervisioned__

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/weights_heights.csv')
```

## Supervisioned

```python
def height_supervisioned_discretization(val) -> str:
    if val < 170:
        return 'SHORT'
    else:
        return 'TALL'

def weight_supervisioned_discretization(val) -> str:
    if val > 70:
        return  'FAT'
    else:
        return 'THIN'
```

```python
df_new = df.copy()
df_new['height'] = df['height'].apply(height_supervisioned_discretization)
df_new['weight'] = df['weight'].apply(weight_supervisioned_discretization)

df_new.sample(frac=0.1)
```

## Unsupervisioned
__Equal-width__ approach using __quantiles__ to discretize.

__Equal-width__ can`t work fine with data in a non normal distribution

```python
df_new = df.copy()

def height_unsupervisioned_discretization(val) -> str:
    if val < df_new['height'].quantile(0.5):
        return 'SHORT'
    else:
        return 'TALL'

def weight_unssupervisioned_discretization(val) -> str:
    if val > df_new['weight'].quantile(0.5):
        return  'FAT'
    else:
        return 'THIN'
```

```python
df_new['height'] = df['height'].apply(height_unsupervisioned_discretization)
df_new['weight'] = df['weight'].apply(weight_unssupervisioned_discretization)

df_new.sample(frac=0.1)
```
