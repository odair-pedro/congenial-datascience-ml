# Binarization
---

Binarization is a tecnique that can be applied above numerical or categorical data.
The main subject is to classify only in two categories (true|false)

In __machine learning__, classification tasks about feature with too many categories can face problems __(Overfitting)__. Here __binarization__ can helps.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/hum-res.csv')
```

```python
df[['Employee Name', 'Department', 'Position']].head()
```

```python
# Getting unique values in the categorical feature
df['Department'].unique()
```

```python
# Binarizing 'Department' categories in 'IT' (true|false)
departments = df['Department']
df['IT'] = (departments == 'IT/IS') | (departments == 'Software Engineering')

df.groupby(['IT', 'Department'])['Employee Name'].count()
```

```python
df.sample(frac=0.05)
```
