### Happiness and money

**Model:** Linear Regression

**Feature:** GDP

**Labels:** Happiness

```python
import pandas as pd

happiness = pd.read_csv('datasets/happiness.csv')
money = pd.read_csv('datasets/money.csv')

df = pd.merge(money, happiness, on='Country', how='inner')

# Convert GDP to float
df['GDP'] = df['GDP'].apply(float)

df.head()
```

```python
df.plot(kind='scatter', x='GDP', y='Happiness')
```

```python
df.head()[['Country', 'GDP']]
```

### Solution

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Note: X (uppercase to matrix) and y (lowercase to array)
X = np.array(df['GDP']).reshape(-1, 1)
y = df['Happiness']

X.shape, y.shape
```

```python
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```

```python
reg.predict(X)
```

```python
# Plotting correlation
import matplotlib.pyplot as plt
prediction_y = reg.predict(X)

plt.scatter(X, y, color='blue')
plt.plot(X, prediction_y, color='red')
plt.show
```

```python
# Predicting Argentina`s happiness
argentina_gpd = [[13588.84]]
reg.predict(argentina_gpd)
```
