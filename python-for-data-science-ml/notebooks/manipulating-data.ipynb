## Using Pandas

```python
import pandas

data_frame = pandas.read_csv('datasets/brain_size.csv', sep=';', na_values='.')
data_frame.head()
```

### Grouping Categorical Variables

```python
groupby_gender = data_frame.groupby('Gender')
for gender, value in groupby_gender['VIQ']:
    print((gender, value.mean()))
```

```python
groupby_gender.mean()
```

### Plotting a dataframe

```python
# Box plots of different columns for each gender
groupby_gender.boxplot(column=['FSIQ', 'VIQ', 'PIQ'])
```

```python
from pandas import plotting
import matplotlib.pyplot as plt

# Scatter matrices for different columns
plotting.scatter_matrix(data_frame[['Weight', 'Height', 'MRI_Count']])
plotting.scatter_matrix(data_frame[['PIQ', 'VIQ', 'FSIQ']])
plt.show()

plotting.scatter_matrix(data_frame[['MRI_Count']])
```

---
Creating from arrays: A pandas.DataFrame can also be seen as a dictionary of 1D ‘series’, eg arrays or lists. If we have 3 numpy arrays

```python
import numpy as np
t = np.linspace(-6, 6, 20)
sin_t = np.sin(t)
cos_t = np.cos(t)

df = pandas.DataFrame({'t': t, 'sin': sin_t, 'cos': cos_t})
df.plot()
```
