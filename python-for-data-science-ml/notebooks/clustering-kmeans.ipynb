## Using `make_blobs`
Source: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=200, n_features=2, centers=10, random_state=40)
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt

x_plt, y_plt = (X[:, 0], X[:, 1])
labels = y

plt.figure()
plt.scatter(x_plt, y_plt, cmap='viridis')
plt.title('Generated blobs')
plt.show()
```

## Implementing K-means

### 4 Clusters

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, n_init=10, max_iter=100).fit(X)
y_means = kmeans.predict(X)

kmeans.labels_, y_means
# labels and y prediction looks be the same
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_means, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()
```

### 8 Clusters

```python
kmeans = KMeans(n_clusters=8, n_init=10, max_iter=100).fit(X)
y_means = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_means, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()
```

## K-means Centroids
https://pythonprogramminglanguage.com/kmeans-clustering-centroid/

```python
colors = ['b', 'g', 'c', 'y']
markers = ['o', 'v', 's', 'd']

K = 4
kmeans_model = KMeans(n_clusters=K).fit(X)

print(kmeans_model.cluster_centers_)
centers = kmeans_model.cluster_centers_

plt.plot()
plt.title('k means centroids')

for i, l in enumerate(kmeans_model.labels_):
    plt.plot(X[i, 0], X[i, 1], color=colors[l], marker=markers[l], ls='None')

plt.scatter(centers[:, 0], centers[:, 1], marker="o", s=100, color='r')
plt.show()
```
