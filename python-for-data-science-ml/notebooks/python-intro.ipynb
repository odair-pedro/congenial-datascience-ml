```python
a = np.arange(10) # 0 .. n-1  (!)
a
```

```python
c = np.linspace(0, 1, 6)   # start, end, num-points
c
```

```python
d = np.linspace(0, 1, 5, endpoint=False)
d
```

```python
a = np.ones((3, 3))  # reminder: (3, 3) is a tuple
a
```

```python
b = np.zeros((2, 2))
b
```

```python
c = np.eye(3)
c
```

### Fancy Indexing

#### Indexing with a mask

```python
np.random.seed(3)
a = np.random.randint(0, 21, 15)
a
```

```python
(a % 3 == 0)
```

```python
mask = (a % 3 == 0)
extract_from_a = a[mask] # or,  a[a%3==0]
extract_from_a           # extract a sub-array with the mask
```

#### Assigning with a mask

```python
a[a % 3 == 0] = -1
a
```

### Indexing with array of integer

```python
a = np.arange(0, 100, 10)
a
```

```python
a[[2, 3, 2, 4, 2]]  # note: [2, 3, 2, 4, 2] is a Python list
```

```python
a[[9, 7]] = -100   # assign using array
a
```

```python
a = np.arange(0, 10, 0.1)
idx = np.array([3, 4])

print(a)
a[idx]
```

```python
a = np.arange(0, 10, 0.1)
idx = np.array([[3, 4], [9, 7]])

print(a)
print(idx.shape)

a[idx]
```

### 2D Arrays plotting

```python
img = np.random.rand(3, 3)
print(img)
plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
```

```python
img = np.random.rand(10, 10)
plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
```

```python
# Sorting values
img = np.sort(img, axis=0)
img = np.sort(img, axis=1)
plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
```

```python
img = np.random.rand(100, 100)
print(img.shape)

plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
```

```python
# Sorting values
img = np.random.rand(100, 100)
img = np.sort(img, axis=0)
plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
plt.show()

img = np.sort(img, axis=1)
plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
plt.show()

print('=================================================')

img = np.random.rand(100, 100)
img = np.sort(img, axis=1)
plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
plt.show()

img = np.sort(img, axis=0)
plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
plt.show()
```

```python
img = np.random.rand(100, 100)
print(img.shape)

plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
```

```python
a1 = np.linspace(0, 2, 10).tolist()
a2 = np.linspace(0, 5, 10).tolist()
img = np.array([a1, a2])

plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
```

```python
arr = [np.linspace(0, x+1, 10).tolist() for x in range(10)]
img = np.array(arr)

plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
```

```python
arr = [np.linspace(0, x+1, 100).tolist() for x in range(100)]
img = np.array(arr)

plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
```

```python
arr1 = [np.linspace(0, x+1, 800, dtype=float) for x in range(400)]
arr2 = [np.linspace(0, 400-x, 800, dtype=float) for x in range(400)]

arr = np.concatenate([arr1, arr2])
img = np.array(arr)
print(img.shape)

plt.imshow(img, cmap=plt.cm.hot)
plt.colorbar()
```

```python
arr = [np.linspace(x, x, 200) for x in range(200)]

img = np.array(arr)
print(img.shape)

plt.imshow(img, cmap=plt.cm.coolwarm)
plt.colorbar()
```

```python
arr1 = [np.linspace(x, x, 400) for x in range(200)]
arr2 = [np.linspace(x, x, 400) for x in reversed(range(200))]

arr = np.concatenate([arr1, arr2])
img = np.array(arr)
print(img.shape)

plt.imshow(img, cmap=plt.cm.coolwarm)
plt.colorbar()
```

```python
# arange can be used as well
np.arange(1, 10, 0.1)
```

```python
ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
