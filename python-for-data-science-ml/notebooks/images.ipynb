```python
import imageio
imageio.imsave('assets/tiny_elephant.jpg', img[::10,::10, ::])

new_img = plt.imread('assets/tiny_elephant.jpg')
print(new_img.ndim)
print(new_img.shape)
print(new_img)

plt.imshow(new_img, interpolation='nearest')
plt.show(block=False)
```
